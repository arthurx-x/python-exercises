### Baixando dados de ativos

1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.

2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

3.  Observando a volatilidade dentro do dia

4. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).

5. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.

6. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
#Importando bibliotecas
import yfinance as yf
import pandas as pd
import plotly.graph_objs as go
import warnings
```

```python
# Ignorar FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

def plotar_amplitude_acoes(tickers, data_inicio, data_fim, figsize=(800, 600)):
    todos_dados = pd.DataFrame()
    traces, trendlines, min_max_points = [], [], []

    # Iterando sobre cada ativo
    for ativo in tickers:
        # Baixando dados do ativo
        dados = yf.download(ativo, start=data_inicio, end=data_fim)
        dados.reset_index(inplace=True)
        # Calculando a amplitude diária do ativo
        dados['Amplitude'] = dados['High'] - dados['Low']
        # Encontrando a data e valor mínimo da amplitude
        min_date = dados.loc[dados['Amplitude'].idxmin()]['Date']
        min_value = dados['Amplitude'].min()
        # Encontrando a data e valor máximo da amplitude
        max_date = dados.loc[dados['Amplitude'].idxmax()]['Date']
        max_value = dados['Amplitude'].max()
        # Criando gráficos de dispersão para os pontos de amplitude
        trace = go.Scatter(x=dados['Date'], y=dados['Amplitude'], mode='lines+markers', name=ativo)
        # Criando linha de tendência da amplitude
        trendline = go.Scatter(x=dados['Date'], y=dados['Amplitude'].rolling(window=30).mean(), mode='lines', name=f'Tendência - {ativo}')
        # Criando pontos para mostrar os valores mínimo e máximo
        min_point = go.Scatter(x=[min_date], y=[min_value], mode='markers', marker=dict(color='red'), name=f'Mínimo - {ativo}')
        max_point = go.Scatter(x=[max_date], y=[max_value], mode='markers', marker=dict(color='green'), name=f'Máximo - {ativo}')
        # Adicionando os gráficos à lista correspondente
        traces.append(trace)
        trendlines.append(trendline)
        min_max_points.extend([min_point, max_point])
        # Concatenando dados do ativo ao DataFrame geral
        todos_dados = pd.concat([todos_dados, dados.set_index('Date')], axis=1)

    # Salvando dados dos ativos em um arquivo CSV
    todos_dados.to_csv('ativos.csv')
    print("Dados dos ativos baixados e salvos com sucesso!")

    # Definindo layout do gráfico
    layout = go.Layout(
        title=dict(text='Variação da Amplitude de Valor do Ativo Dentro do Dia', font=dict(color='#FFFFFF', size=20)),
        xaxis=dict(title='Data', tickangle=45, tickfont=dict(color='#FFFFFF', size=14), titlefont=dict(color='#FFFFFF', size=16)),
        yaxis=dict(title='Amplitude de Variação', tickfont=dict(color='#FFFFFF', size=14), titlefont=dict(color='#FFFFFF', size=16)),
        showlegend=True,
        plot_bgcolor='rgb(0, 0, 0)',
        paper_bgcolor='rgb(0, 0, 0)',
        legend=dict(font=dict(color='#FFFFFF')),
        width=figsize[0],
        height=figsize[1],
        xaxis_showgrid=False,
        yaxis_showgrid=False
    )

    # Criando figura do gráfico
    fig = go.Figure(data=traces + trendlines + min_max_points, layout=layout)
    fig.show()

tickers = ['NVDA', 'MSFT', 'AAPL']
data_inicio = '2022-01-01'
data_fim = '2024-02-09'
figsize = (1000, 500)

plotar_amplitude_acoes(tickers, data_inicio, data_fim, figsize)
```
