### 04 - Números primos
> Um número **N** é primo se e somente se é divisível por 1, -1, por **N** e por -**N**.  

Escreva um script que verifica se ```N``` é um número primo, verificando se ```N``` é divisível por todos os números de ```1``` a ```N-1```. Você vai precisar usar alguma ferramenta de *loop* que você aprendeu para isto. No final, devolva uma mensagem no output indicando se o número é primo ou não.

````python
# Define a variável N
N = 47

# Inicia a variável primo como True
primo = True

# Percorre os números de 2 até N-1
for i in range(2, N):
    # Verifica se N é divisível por i usando o resto da divisão
    if N % i == 0:
        # Se for divisível, muda a variável primo para False e sai do loop
        primo = False
        break

# Verifica se a variável primo é True
if primo:
    # Se for True, imprime uma mensagem dizendo que N é primo
    print(f"{N} é um número primo.")
else:
    # Se for False, imprime uma mensagem dizendo que N não é primo
    print(f"{N} não é um número primo.")
````

### 06 - Peso ideal 1
O IMC (índice de massa corpórea) é um indicador de saúde mais bem aceito que o peso. Ele é calculado como:

$$ IMC = \dfrac{peso}{altura^2}$$

Segundo a OMS, valores *normais* são entre 18.5 e 24.9.

Sua tarefa é encontrar o ponto médio dessa faixa.

````python
# Define uma lista com os limites inferior e superior do IMC ideal
imc_ideal = [18.5, 24.9]

# Calcula a média dos limites usando a função np.mean
imc_ideal_medio = np.mean(imc_ideal)

# Imprime o valor do IMC ideal médio
print(imc_ideal_medio)
````

### 07 - Peso ideal 2
Recebendo um valor de altura, encontre o peso '*ideal*' dessa pessoa, que fornece o IMC encontrado acima

````python
# Define a variável altura
altura = 1.70

# Converte a variável altura em um float
altura = float(altura)

# Define uma lista com os limites inferior e superior do IMC ideal
imc_ideal = [18.5, 24.9]

# Converte a lista em um array do numpy
imc_ideal = np.array(imc_ideal)

# Calcula a média dos limites usando a função np.mean
imc_ideal_medio = np.mean(imc_ideal)

# Calcula o peso ideal usando a fórmula do IMC: peso = IMC * altura**2
peso_ideal = float(imc_ideal_medio) * altura**2

# Imprime o valor do peso ideal para a altura dada, formatando o resultado com duas casas decimais
print(f"O peso ideal para uma altura de {altura}m é {peso_ideal:.2f}kg")
````

### 10 - Peso ideal 5
Dada uma lista de **listas** - cada elemento da lista é uma **lista** contendo altura e peso de um paciente, adicione mais um elemento à lista de cada paciente contendo o IMC do paciente. Verifique também se é 'baixo', 'normal' ou 'alto' segundo os padrões da OMS em que normal é entre 18.5 e 24.9.

Reflexão: por que no problema anterior temos que criar uma nova lista, e não podemos adicionar os dados de cada indivíduo à tupla?

````python
# Array de altura e peso para cada paciente
altura_peso = np.array([[1.80, 90], [1.65, 75], [1.91, 70]])

# Cálculo do IMC para cada paciente
imc = altura_peso[:, 1] / altura_peso[:, 0] ** 2
imc_arredondado = np.round(imc, 2)

# Adicionando colunas de IMC e Classificação ao array
altura_peso = np.column_stack((altura_peso, imc_arredondado))
altura_peso = np.column_stack((altura_peso, np.select([imc < 18.5, (18.5 <= imc) & (imc <= 24.9), imc > 24.9], ['baixo', 'normal', 'alto'])))

# Cabeçalho para a tabela
headers = ["Altura", "Peso", "IMC", "Classificação"]

# Impressão da tabela utilizando o módulo tabulate
print(tabulate(altura_peso, headers=headers))
````

### 4) É muito comum precisarmos identificar valores especiais e darmos tratamento a eles quer seja alterando-os quer seja descartando-os. 

O trecho de código abaixo gera um ndarray com números pseudo aleatórios. Considere que para efeitos do estudo que virá, devemos desconsiderar valores iguais a zero. Sendo assim:

i) crie um objeto ```bool_zero``` que traga uma sequencia de booleanos do mesmo tamanho que o objeto poi, e que vale ```True``` quando o valor de ```poi``` é zero, e ```False``` caso contrário.

ii) Conte quantos valores zero existem. Lembre-se de que no final das contas, ```True``` vale 1 para o Python, e ```False``` vale zero, então uma boa dica seria usar a função ```sum()```.

iii) Utilize a indexação booleana que você aprendeu para criar uma variável ```poi_nao_zero``` que aponta para os elementos de ```poi``` diferentes de zero. Dica: você vai pode inverter os elementos do objeto que criou em *ii)* ou escrever a comparação adequada.

````python
# Objeto 'poi' gerado com distribuição de Poisson, com semente para reprodutibilidade
np.random.seed(1234)
poi = np.random.poisson(3, 100)

# Verifica quais valores na array 'poi' são iguais a zero
bool_zero = (poi == 0)
````

````python
# Cria o objeto bool_zero através do operador ==
bool_zero = (poi == 0)
print(bool_zero)
````

````python
# Conta a quantidade de zeros (ou soma os elementos de bool_zero)
num_zeros = np.sum(bool_zero)
print(num_zeros)
````

````python
# Utiliza a indexação booleana para criar uma seleção de não-zeros
poi_nao_zero = poi[~bool_zero]

# Impressão dos resultados
print("Objeto bool_zero:")
print(bool_zero)
print("\nNúmero de valores zero:", num_zeros)
print("\nObjeto poi_nao_zero:")
print(poi_nao_zero)
````
