```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm, t
from prettytable import PrettyTable
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Probabilidade de um pacote ter mais que 15,2 kg
probabilidade = 1 - norm.cdf(15.2, loc=15, scale=0.1)
print(f"Probabilidade de um pacote ter mais que 15,2 kg: {probabilidade:.4f}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Percentil de ordem 95% para o peso dos pacotes
percentil_95 = norm.ppf(0.95, loc=15, scale=0.1)
print(f"Percentil de ordem 95% para o peso dos pacotes: {percentil_95:.4f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# Probabilidade de ser menor que valores em Y
Y = np.linspace(-6, 6, 200)

# Normal Padrão
prob_normal = norm.cdf(Y)

# t-student com 5 graus de liberdade
prob_student = t.cdf(Y, df=5)

# Criar tabela
table = PrettyTable()
table.field_names = ["Valor de X", "Normal Padrão", "t-Student (5 gl)"]

# Adicionar dados à tabela
table.add_row(["-3", f"{prob_normal[25]:.4f}", f"{prob_student[25]:.4f}"])
table.add_row(["0", f"{prob_normal[100]:.4f}", f"{prob_student[100]:.4f}"])
table.add_row(["2", f"{prob_normal[150]:.4f}", f"{prob_student[150]:.4f}"])

# Imprimir tabela
print("Probabilidades P(X < valor)")
print(table)
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Visualização gráfica
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(Y, prob_normal, label='Normal Padrão')
ax.plot(Y, prob_student, label='t-student (5 gl)')
ax.set_xlabel('Valores de Y')
ax.set_ylabel('Probabilidade Acumulada')
ax.legend()
plt.show()
```

5. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Percentis (inversa da CDF) para valores em P
P = np.linspace(0, 1, 20)

# Normal Padrão
percentis_normal = norm.ppf(P)

# t-student com 5 graus de liberdade
percentis_student = t.ppf(P, df=5)

# Criar tabela
table = PrettyTable()
table.field_names = ["Probabilidade Acumulada", "Normal Padrão", "t-Student (5 gl)"]

# Adicionar dados à tabela
percentis = [0.05, 0.5, 0.95]
for p in percentis:
    table.add_row([f"{p*100:>3.0f}%", f"{norm.ppf(p):>7.4f}", f"{t.ppf(p, df=5):>7.4f}"])

# Imprimir tabela
print("Percentis (inversa da CDF)")
print(table)
```

6. Faça um gráfico com os dados do item anterior.

```python
# Visualização gráfica
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(P, percentis_normal, label='Normal Padrão')
ax.plot(P, percentis_student, label='t-student (5 gl)')
ax.set_xlabel('Probabilidade Acumulada')
ax.set_ylabel('Valores dos Percentis')
ax.legend()
plt.tight_layout()
plt.show()
```
