#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
import numpy as np
import warnings
```

```python
# Carrega os dados de nascidos do arquivo 'SINASC_RO_2019.csv' para um DataFrame
nascidos = pd.read_csv('SINASC_RO_2019.csv')

# Exibe as primeiras linhas do DataFrame para visualização inicial
nascidos.head()
```

```python
# Imprime os nomes das colunas do DataFrame 'dados_nascidos'
print(nascidos.columns)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Configura o tamanho da figura
warnings.simplefilter(action='ignore', category=FutureWarning)
plt.figure(figsize=(15, 20))

# Subplot 1: Distribuição de APGAR1
plt.subplot(5, 1, 1)
sns.countplot(data=nascidos, x='APGAR1')
plt.title('Distribuição de APGAR1')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

# Subplot 2: Distribuição de APGAR5
plt.subplot(5, 1, 2)
sns.countplot(data=nascidos, x='APGAR5')
plt.title('Distribuição de APGAR5')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

# Subplot 3: Distribuição de SEXO
plt.subplot(5, 1, 3)
sns.countplot(data=nascidos, x='SEXO')
plt.title('Distribuição de SEXO')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

# Subplot 4: Distribuição de GESTACAO
plt.subplot(5, 1, 4)
sns.countplot(data=nascidos, x='GESTACAO')
plt.title('Distribuição de GESTACAO')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

# Subplot 5: Distribuição de ESTCIVMAE
plt.subplot(5, 1, 5)
sns.countplot(data=nascidos, x='ESTCIVMAE')
plt.title('Distribuição de ESTCIVMAE')
plt.yscale("log")
plt.xlabel('ESTCIVMAE')

# Ajusta o layout para evitar sobreposição
plt.tight_layout()

# Exibe o gráfico
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Configura o tamanho da figura
plt.figure(figsize=(12, 6))

# Cria o gráfico de barras
ax = sns.barplot(data=nascidos, x='CONSULTAS', y='APGAR5', color='blue', ci=None)

# Adiciona título e rótulos aos eixos
ax.set(title='APGAR5 Médio por Número de Consultas de Pré-natal', xlabel='Número de Consultas', ylabel='Média de APGAR5')

# Exibe o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Configurando o tamanho da figura
plt.figure(figsize=(15, 30))

# Função para simplificar a criação dos subplots
def criar_subplot(posicao, feature, titulo):
    plt.subplot(4, 1, posicao)
    sns.countplot(data=nascidos, x=feature)
    plt.title(f'Distribuição de {titulo}')
    plt.yscale("log")

# Criando os subplots chamando a função
criar_subplot(1, 'SEXO', 'Sexo')
criar_subplot(2, 'GESTACAO', 'Duração da Gestação')
criar_subplot(3, 'ESTCIVMAE', 'Estado Civil da Mãe')
criar_subplot(4, 'CONSULTAS', 'Número de Consultas Pré-natal')

# Ajustando o layout e exibindo o gráfico
plt.tight_layout()
plt.show()
```

# Atividade 03

#### 1) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Definição dos intervalos e rótulos para categorização do APGAR5
bins = [-1, 3, 5, 7, 10]
labels = ['asfixia grave', 'asfixia moderada', 'asfixia leve', 'normal']

# Criação da coluna 'APGAR5_categoria' com base na categorização do APGAR5
nascidos['APGAR5_categoria'] = pd.cut(nascidos['APGAR5'], bins=bins, labels=labels)

# Exibição das primeiras 15 linhas das colunas 'APGAR5' e 'APGAR5_categoria'
print(nascidos[['APGAR5', 'APGAR5_categoria']].head(15))
```

#### 2) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# Criação do gráfico de contagem categorizado por 'APGAR5_categoria', 'SEXO' e 'GESTACAO'
g = sns.catplot(data=nascidos, x='APGAR5_categoria', hue='SEXO', col='GESTACAO', col_wrap=2, kind='count')

# Configuração da escala logarítmica no eixo y e definição dos rótulos e ticks
g.set(yscale='log', yticks=[1, 10, 100, 1000, 10000], yticklabels=['1', '10', '100', '1000', '10000'])

# Configuração do rótulo do eixo y
g.set_ylabels('Count')

# Exibição do gráfico
plt.show()
```

####  3) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Criação da tabela cruzada entre 'GESTACAO' e 'APGAR5_categoria'
tabela_cruzada = pd.crosstab(nascidos['GESTACAO'], nascidos['APGAR5_categoria'])

# Exibição da tabela cruzada
print(tabela_cruzada)
```

####  4) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Criação da tabela cruzada entre 'GESTACAO' e 'APGAR5_categoria'
tabela_cruzada = pd.crosstab(nascidos['GESTACAO'], nascidos['APGAR5_categoria'])

# Criação da tabela percentual em relação às linhas
tabela_percentual_linha = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)

# Exibição da tabela percentual em relação às linhas
print(tabela_percentual_linha)
```

####  5) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Criação da tabela cruzada entre 'GESTACAO' e 'APGAR5_categoria'
tabela_cruzada = pd.crosstab(nascidos['GESTACAO'], nascidos['APGAR5_categoria'])

# Criação da tabela percentual em relação às linhas
tabela_percentual_linha = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)

# Gráfico de barras empilhadas do percentual em relação às linhas
tabela_percentual_linha.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 8))

# Configurações do título e rótulos dos eixos
plt.title('Distribuição APGAR5 por Tempo de Gestação')
plt.xlabel('Tempo de Gestação')
plt.ylabel('Percentual')

# Configuração da legenda
plt.legend(title='Categoria APGAR5', bbox_to_anchor=(1.05, 1), loc='upper left')

# Exibição do gráfico
plt.show()
```
